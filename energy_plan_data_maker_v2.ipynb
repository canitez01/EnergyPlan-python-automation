```python
#working unique combinations, https://www.geeksforgeeks.org/python-program-to-get-all-unique-combinations-of-two-lists/#:~:text=The%20unique%20combination%20of%20two,elements%20of%20the%20second%20list.&text=Method%201%20%3A%20Using%20permutation(),and%20initialize%20list_1%20and%20list_2.
import itertools
from itertools import product
 
# initialize lists
list_1 = ["d_1","d_2","d_3","s_p"]
list_2 = ["s_a","s_b","s_c"]
 
# create empty list to store the combinations
unique_combinations = []
 
# Extract Combination Mapping in two lists
# using zip() + product()
unique_combinations = list(list(zip(list_1, element))
                           for element in product(list_2, repeat = len(list_1)))

loop_list=[sum(m,()) for m in unique_combinations]
```

```python
from copy import deepcopy
```

```python
#for every combination create scenario inputs
df_synth=deepcopy(df)
noms=[]
for m in loop_list:
    print(m)
    #d_1 dünyası
    z=df_codes.loc[m[0], ["name",m[1]]]
    itemDict1 = {item[0]: item[1] for item in z.values.tolist()}
    for key,values in itemDict1.items():
        df_synth.loc[df_synth.loc[(df_synth[0]==key)].index[0]+1]=values
    #d_2 dünyası
    t=df_codes.loc[m[2], ["name",m[3]]]
    itemDict2 = {item[0]: item[1] for item in t.values.tolist()}
    for key,values in itemDict2.items():
        df_synth.loc[df_synth.loc[(df_synth[0]==key)].index[0]+1]=values
    #d_3 dünyası
    y=df_codes.loc[m[4], ["name",m[5]]]
    itemDict3 = {item[0]: item[1] for item in y.values.tolist()}
    for key,values in itemDict3.items():
        df_synth.loc[df_synth.loc[(df_synth[0]==key)].index[0]+1]=values
    #s_p dünyası
    u=df_codes.loc[m[6], ["name",m[7]]]
    itemDict4 = {item[0]: item[1] for item in u.values.tolist()}
    for key,values in itemDict4.items():
        df_synth.loc[df_synth.loc[(df_synth[0]==key)].index[0]+1]=values
    nom="-".join(m)+".txt"
    noms.append(nom)
    df_synth.to_csv("C:/Users/User/Desktop/Kaynaklar Master/EnergyPlan/ZipEnergyPLAN160/ZipEnergyPLAN160/energyPlan Data/Data/deneme"+"/"+nom, index=False, header=False)
    
```
